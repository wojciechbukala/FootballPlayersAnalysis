# Laboratorium 3 — t-SNE i UMAP

Krótki notebook dla chętnych pokazujący redukcję wymiarów na małym podzbiorze `digits`.


## Przydatne materiały
- `t-SNE`: https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html
- `UMAP`: https://umap-learn.readthedocs.io/en/latest/


```python
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid")
```

```python
digits = load_digits(as_frame=True)
digits_df = digits.frame.sample(n=300, random_state=42)
X_digits = digits_df.drop(columns=['target'])
y_digits = digits_df['target'].astype(str)
X_tsne = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto').fit_transform(X_digits)
try:
    import umap.umap_ as umap
    X_umap = umap.UMAP(n_components=2, random_state=42).fit_transform(X_digits)
except Exception as exc:
    X_umap = None
    umap_error = str(exc)
```

```python
if X_umap is None:
    fig, ax = plt.subplots(figsize=(7, 5))
    sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y_digits, palette="tab10", s=35, ax=ax, legend=False)
    ax.set_title("t-SNE na digits")
    plt.tight_layout()
    print("UMAP nie jest dostępny w tym środowisku.")
    print("Spróbuj np.: uv add umap-learn setuptools<81")
    print(umap_error)
else:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=y_digits, palette="tab10", s=35, ax=axes[0], legend=False)
    axes[0].set_title("t-SNE na digits")
    sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=y_digits, palette="tab10", s=35, ax=axes[1], legend=False)
    axes[1].set_title("UMAP na digits")
    plt.tight_layout()
```
