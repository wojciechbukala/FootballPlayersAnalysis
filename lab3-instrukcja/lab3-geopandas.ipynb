## Przydatne materiały
- `GeoPandas`: https://geopandas.org/en/stable/docs.html
- `geodatasets`: https://geodatasets.readthedocs.io/en/latest/


```python
import geopandas as gpd
import matplotlib.pyplot as plt
import geodatasets
```

```python
path = geodatasets.get_path("naturalearth.land")
land = gpd.read_file(path)
land.head()
```

```python
land.shape
```

```python
land.plot(figsize=(10, 6), color="lightgreen", edgecolor="black")
plt.title("GeoPandas: lądy świata (Natural Earth)")
plt.show()
```
